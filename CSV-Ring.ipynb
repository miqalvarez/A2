# A2-Ring

### Read data

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Read the different txt files
df1 = pd.read_csv('A2-ring\A2-ring-merged.txt', sep="\t", header=None)
df2 = pd.read_csv('A2-ring\A2-ring-separable.txt', sep="\t", header=None)
df3 = pd.read_csv('A2-ring\A2-ring-test.txt', sep="\t", header=None)
```

### Preprocess

```python
# Shuffle the data and save to csv
df1 = df1.sample(frac=1)
df2 = df2.sample(frac=1)
df3 = df3.sample(frac=1)

df1.to_csv('A2-ring\A2-ring-merged.csv', index=False, header=False)
df2.to_csv('A2-ring\A2-ring-separable.csv', index=False, header=False)
df3.to_csv('A2-ring\A2-ring-test.csv', index=False, header=False)
```

### Plot the data

```python
plt.scatter(df1[0], df1[1], c=df1[2])
plt.title("A2-ring-merged")
plt.show()

plt.scatter(df2[0], df2[1], c=df2[2])
plt.title("A2-ring-separable")
plt.show()

plt.scatter(df3[0], df3[1], c=df3[2])
plt.title("A2-ring-test")
plt.show()
```
